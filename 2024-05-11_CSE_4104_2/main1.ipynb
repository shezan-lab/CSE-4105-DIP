```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
frequency = 5  # Frequency of the sine wave in Hz
amplitude = 1  # Amplitude of the sine wave
duration = 1   # Duration of the signal in seconds
sampling_rate = 1000  # Sampling rate in samples per second

# Time vector
t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)

# Generate the sine wave
analog_signal = amplitude * np.cos(2 * np.pi * frequency * t)

# Plot the analog signal
plt.plot(t, analog_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Analog Signal (Cosine Wave)')
plt.show()
```

```python
# Sample the analog signal
sampling_indices = np.arange(0, len(analog_signal), sampling_rate)
sampled_signal = analog_signal[sampling_indices]

# Quantization (for simplicity, let's assume uniform quantization)
num_bits = 8  # Number of bits for quantization
quantization_levels = 2 ** num_bits
quantized_signal = np.round(sampled_signal * (quantization_levels - 1)) / (quantization_levels - 1)

# Plot the sampled and quantized signals
plt.stem(sampled_signal, label='Sampled Signal')
plt.plot(quantized_signal, 'r-', label='Quantized Signal')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.title('Sampled and Quantized Signals')
plt.legend()
plt.show()
```

```python
# Unit Impulse/Sample
def unit_impulse(n):
    return 1 if n == 0 else 0

# Unit Step
def unit_step(n):
    return 1 if n >= 0 else 0

# Unit Ramp
def unit_ramp(n):
    return n if n >= 0 else 0

```

```python
# Shifting
def shift_signal(signal, shift):
    return np.roll(signal, shift)

# Scaling
def scale_signal(signal, scale):
    return signal * scale

# Reversal
def reverse_signal(signal):
    return np.flip(signal)
```
